```python
def visualize_dataframe(dataframe):
    for column in dataframe.columns:
        column_type = dataframe[column].dtype

        if column_type == 'object':
            value_counts = dataframe[column].value_counts()
            value_counts.plot.pie(autopct='%1.1f%%', startangle=90)
            plt.title(f'{column} - –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞')
            plt.show()

        elif column_type in ['int64', 'float64'] and column != 'id':
            fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

            dataframe[column].plot.hist(bins=25, ax=axes[0])
            axes[0].set_title(f'{column} - –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞')
            axes[0].set_xlabel(column)
            axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

            sns.boxplot(x=dataframe[column], ax=axes[1])
            axes[1].set_title(f'{column} - Boxplot')
            axes[1].set_xlabel(column)

            plt.show()
            
            print(dataframe[column].describe())
```

```python
RANDOM_STATE = 97
TEST_SIZE = 0.25
X_train, X_test, y_train, y_test = train_test_split(
    market_full.drop(
        [
             '–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'
        ], 
        axis=1
    ),
    market_full['–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = market_full['–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å']
)

# —Å–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–∫–∏ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
ohe_columns = [
    '–†–∞–∑—Ä–µ—à–∏—Ç—å —Å–æ–æ–±—â–∞—Ç—å', '–ü–æ–ø—É–ª—è—Ä–Ω–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è'
]
ord_columns = [
    '–¢–∏–ø —Å–µ—Ä–≤–∏—Å–∞', 
]
num_columns = ['–ú–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_6_–º–µ—Å', '–ú–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_—Ç–µ–∫_–º–µ—Å', '–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å', '–ê–∫—Ü–∏–æ–Ω–Ω—ã–µ_–ø–æ–∫—É–ø–∫–∏',
                      '–°—Ä–µ–¥–Ω–∏–π_–ø—Ä–æ—Å–º–æ—Ç—Ä_–∫–∞—Ç–µ–≥–æ—Ä–∏–π_–∑–∞_–≤–∏–∑–∏—Ç', '–ù–µ–æ–ø–ª–∞—á–µ–Ω–Ω—ã–µ_–ø—Ä–æ–¥—É–∫—Ç—ã_—à—Ç—É–∫_–∫–≤–∞—Ä—Ç–∞–ª',
                      '–û—à–∏–±–∫–∞_—Å–µ—Ä–≤–∏—Å–∞', '–°—Ç—Ä–∞–Ω–∏—Ü_–∑–∞_–≤–∏–∑–∏—Ç',
                      '–≤—ã—Ä—É—á–∫–∞_—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü', '–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü', '–º–∏–Ω—É—Ç_–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü',
                      '–º–∏–Ω—É—Ç_—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü']

# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
        )
    ]
)

# c–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# SimpleImputer + OE
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['–ø—Ä–µ–º–∏—É–º', '—Å—Ç–∞–Ω–¥–∞—Ä—Ç']
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)

data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),  # –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        ('ord', ord_pipe, ord_columns),  # –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ—Ä—è–¥–∫–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        ('num', MinMaxScaler(), num_columns) 
    ], 
    remainder='passthrough'
)
print(data_preprocessor)
```

- GridSearch/RandomizedSearch —ç—Ç–æ –∫–ª–∞—Å—Å–∏–∫–∞, –Ω–æ –µ—Å—Ç—å –∏ –ø–æ–∏–Ω—Ç–µ—Ä–µ—Å–Ω–µ–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –ø–æ–¥–±–æ—Ä–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –º–æ–∂–µ—à—å [–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è](https://www.freecodecamp.org/news/hyperparameter-optimization-techniques-machine-learning/). –í—ã–¥–µ–ª—é –æptuna, –æ—á–µ–Ω—å –º–Ω–æ–≥–æ –ø–ª—é—Å–æ–≤, –ø—Ä–∏—á–µ–º –∏–∑—É—á–µ–Ω–∏–µ –º–æ–∂–Ω–æ –Ω–∞—á–∞—Ç—å —Å [**OptunaSearchCV**](https://optuna.readthedocs.io/en/stable/reference/generated/optuna.integration.OptunaSearchCV.html). –ò–Ω—Ç–µ—Ä—Ñ–µ–π—Å –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ —Ç–∞–∫–æ–π –∂–µ, –∫–∞–∫ —É GridSearch/RandomizedSearch, –ø–æ—ç—Ç–æ–º—É –æ—á–µ–Ω—å –ª–µ–≥–∫–æ –Ω–∞—á–∞—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è

    
- –≤ –Ω–∞—à–µ–º –ø—Ä–æ–µ–∫—Ç–µ –º—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ pipeline —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º–∏ —Ñ—É–Ω–∫—Ü–∏—è–º–∏ sklearn (StandardScaler, MinMaxScaler –∏—Ç–¥ –∏—Ç–ø), –Ω–æ –≤–µ–¥—å —É –Ω–∞—Å –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤–æ—è —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö. –ù–∞–ø—Ä–∏–º–µ—Ä –º—ã —Ö–æ—Ç–∏–º –∑–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ —Å—Ä–µ–¥–Ω–µ–π –≤ –∫–∞–∫–æ–π —Ç–æ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–µ, –Ω–æ –¥–æ–ª–∂–Ω—ã —É—á–µ—Å—Ç—å –ª–æ–≥–∏–∫—É –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã–µ –∏–∑ train (–¥–µ–ª–∞–µ–º .fit), –∞ –∑–∞—Ç–µ–º "–æ–±—É—á–µ–Ω–Ω—É—é –Ω–∞ train —Ñ—É–Ω–∫—Ü–∏—é –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è" –∏—Å–ø–æ–ª—å–∑—É–µ–º –Ω–∞ test/valid (–¥–µ–ª–∞–µ–º .transform), –ø–æ —Å—Å—ã–ª–∫–∞–º –º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∫–∞–∫ —ç—Ç–æ –º–æ–∂–Ω–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å [–°—Å—ã–ª–∫–∞ 1](https://dzen.ru/media/id/5ee6f73b7cadb75a66e4c7e3/sozdanie-polzovatelskih-preobrazovatelei-dannyh-62b2a9a80e49941961ffc7a2),
[–°—Å—ã–ª–∫–∞ 2](https://towardsdatascience.com/pipelines-custom-transformers-in-scikit-learn-the-step-by-step-guide-with-python-code-4a7d9b068156)

–û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ —á—Ç–æ –º—ã —Å–æ–∑–¥–∞–µ–º –¥–ª—è —ç—Ç–æ–≥–æ –∫–ª–∞—Å—Å, –Ω–∞—Å–ª–µ–¥—É—è –∫–ª–∞—Å—Å—ã TransformerMixin –∏ BaseEstimator, —á—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ –ø—Ä–∏–º–µ–Ω—è—Ç—å  .fit(X_train), .predict(X_test)    

-  –±—ã–≤–∞–µ—Ç –ø–æ–ª–µ–∑–Ω–æ –∑–∞–º–µ—Ä–∏—Ç—å –≤—Ä–µ–º—è –∏—Å–ø–æ–ª–Ω–µ–Ω–∏—è –∫–æ–¥–∞ –≤ —è—á–µ–π–∫–µ, —ç—Ç–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å –ø–æ–º–æ—â—å—é –º–∞–≥–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ %%time

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],  # –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ probability=True –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è predict_proba
        'models__C': [0.1, 1, 10],  # –°–ø–∏—Å–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ C
        'models__kernel': ['linear', 'rbf'],  # –°–ø–∏—Å–æ–∫ —è–¥–µ—Ä
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –ø–æ –∫—Ä–æ—Å—Å –≤–∞–ª–∏–¥–∞—Ü–∏–∏:', randomized_search.best_score_)

# –ø—Ä–æ–≤–µ—Ä—å—Ç–µ —Ä–∞–±–æ—Ç—É –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
# —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
y_test_pred_proba = randomized_search.predict_proba(X_test)

# –í–∑—è—Ç—å –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∫–ª–∞—Å—Å —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é
y_test_pred = y_test_pred_proba[:, 1]

# –†–∞—Å—Å—á–∏—Ç–∞—Ç—å ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc_score(y_test, y_test_pred)}')
```

- —ç—Ç–æ –Ω–µ–≤–µ—Ä–Ω–æ .best_score_ —ç—Ç–æ –ú–µ—Ç—Ä–∏–∫–∞ –ø–æ –∫—Ä–æ—Å—Å –≤–∞–ª–∏–¥–∞—Ü–∏–∏, –º—ã –Ω–µ –≤—ã–±–∏—Ä–∞–µ–º –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π –≤—ã–±–æ—Ä–∫–µ, –ø–æ—Ç–æ–º—É —á—Ç–æ –æ–Ω–∞ –Ω–∏—á–µ–≥–æ –Ω–µ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç. –ù–∞–ø—Ä–∏–º–µ—Ä –µ—Å–ª–∏ –≤–∑—è—Ç—å –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–π –≥–ª—É–±–∏–Ω—ã, —Ç–æ –º–µ—Ç—Ä–∏–∫–∞ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞ 100%, –Ω–æ —ç—Ç–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–ø–æ–º–∏–Ω–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö, –∞ –Ω–µ –æ–±–æ–±—â–µ–Ω–∏—è. –†–∞–Ω—å—à–µ –±—ã–ª –ø—Ä–æ–µ–∫—Ç "–≤–µ–¥–µ–Ω–∏—è –º–∞—à–∏–Ω–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ", –≥–¥–µ —á—ë—Ç–∫–æ –æ–± —ç—Ç–æ–º —Ä–∞—Å—Å–∫–∞–∑—ã–≤–∞–ª–æ—Å—å - –æ–±—É—á–∞–µ–º—Å—è –Ω–∞ train, –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ø–æ–¥–±–∏—Ä–∞–µ–º –ø–æ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ (–≤ —Å–ª—É—á–∞–∏ GS –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π), –∞ –∑–∞—Ç–µ–º –ø—Ä–æ–≤–æ–¥–∏–º —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ. –í –ø—Ä–µ–∂–Ω–µ–º –ø—Ä–æ–µ–∫—Ç–µ —Å—Ç—É–¥–µ–Ω—Ç—ã –≤—Ä—É—á–Ω—É—é —Ä–∞–∑–±–∏–≤–∞–ª–∏ –Ω–∞ —Ç—Ä–∏ –≤—ã–±–æ—Ä–∫–∏ –∏ –≤ —Ü–∏–∫–ª–µ –ø–æ–¥–±–∏—Ä–∞–ª–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∞ —Å–µ–π—á–∞—Å —Å—Ä–∞–∑—É GS, –Ω–æ –æ–Ω –¥–µ–ª–∞–µ—Ç —Ç–æ –∂–µ —Å–∞–º–æ–µ: —Ç—ã –ø–æ–¥–∞—ë—à—å –Ω–µ–≥–æ X_train, –∏ –æ–Ω –≤–Ω—É—Ç—Ä–∏ —Å–µ–±—è —Ä–∞–∑–±–∏–≤–∞–µ—Ç –µ–≥–æ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é (–ø—Ä–∏—á–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑)... –í –æ–±—â–µ–º —Ç–∞–∫–æ–π –º–æ–º–µ–Ω—Ç–∏–∫ –ú–æ–∂–µ—à—å –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å [—Ç—É—Ç](https://towardsdatascience.com/why-do-we-need-a-validation-set-in-addition-to-training-and-test-sets-5cf4a65550e0) —Å –ø–æ–º–æ—â—å—é VPN –æ –ª–æ–≥–∏–∫–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫   



</font><font color='Blue'><b>–í –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å –∏—Å–ø–æ–ª—å–∑—É—é ROC AUC, –ø–æ—Ç–æ–º—É —á—Ç–æ –æ–Ω–∞ —è–≤–ª—è–µ—Ç—Å—è —É—Å—Ç–æ–π—á–∏–≤–æ–π –º–µ—Ç—Ä–∏–∫–æ–π –∫ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –∫–ª–∞—Å—Å–∞–º. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ ROC AUC —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤ —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ –∫–ª–∞—Å—Å–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–µ–æ–¥–∏–Ω–∞–∫–æ–≤—ã, –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç F1-–º–µ—Ä—ã,–Ω–∞–ø—Ä–º–µ—Ä, –∫–æ—Ç–æ—Ä–∞—è –º–æ–∂–µ—Ç –±—ã—Ç—å –≤–æ—Å–ø—Ä–∏–∏–º—á–∏–≤–∞ –∫ –¥–∏—Å–±–∞–ª–∞–Ω—Å—É –∫–ª–∞—Å—Å–æ–≤.</b></font><br>

–°–æ–≤–µ—Ç ü§î:



–î–∏—Å–±–∞–ª–∞–Ω—Å —É –Ω–∞—Å —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –º–µ—Ç—Ä–∏–∫–∞ accuracy, —É f1 –Ω–µ—Ç —Å —ç—Ç–∏–º –ø—Ä–æ–±–ª–µ–º.  –í –æ—Å—Ç–∞–ª—å–Ω–æ–º –≤–µ—Ä–Ω–æ


 ## –®–∞–≥ 7.–ê–Ω–∞–ª–∏–∑ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
RANDOM_STATE = 97
TEST_SIZE = 0.25
X_train, X_test, y_train, y_test = train_test_split(
    market_full.drop(['–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'], axis=1),
    market_full['–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'],
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=market_full['–ü–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å']
)

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –∫–æ–ª–æ–Ω–æ–∫ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
ohe_columns = ['–†–∞–∑—Ä–µ—à–∏—Ç—å —Å–æ–æ–±—â–∞—Ç—å', '–ü–æ–ø—É–ª—è—Ä–Ω–∞—è_–∫–∞—Ç–µ–≥–æ—Ä–∏—è']
ord_columns = ['–¢–∏–ø —Å–µ—Ä–≤–∏—Å–∞']
num_columns = ['–ú–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_6_–º–µ—Å', '–ú–∞—Ä–∫–µ—Ç_–∞–∫—Ç–∏–≤_—Ç–µ–∫_–º–µ—Å', '–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å', '–ê–∫—Ü–∏–æ–Ω–Ω—ã–µ_–ø–æ–∫—É–ø–∫–∏',
               '–°—Ä–µ–¥–Ω–∏–π_–ø—Ä–æ—Å–º–æ—Ç—Ä_–∫–∞—Ç–µ–≥–æ—Ä–∏–π_–∑–∞_–≤–∏–∑–∏—Ç', '–ù–µ–æ–ø–ª–∞—á–µ–Ω–Ω—ã–µ_–ø—Ä–æ–¥—É–∫—Ç—ã_—à—Ç—É–∫_–∫–≤–∞—Ä—Ç–∞–ª',
               '–û—à–∏–±–∫–∞_—Å–µ—Ä–≤–∏—Å–∞', '–°—Ç—Ä–∞–Ω–∏—Ü_–∑–∞_–≤–∏–∑–∏—Ç', 
               '–≤—ã—Ä—É—á–∫–∞_—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü', '–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü', '–º–∏–Ω—É—Ç_–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü',
               '–º–∏–Ω—É—Ç_—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü']

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
ohe_imputer = SimpleImputer(strategy='most_frequent')
X_train_ohe = ohe_imputer.fit_transform(X_train[ohe_columns])
X_test_ohe = ohe_imputer.transform(X_test[ohe_columns])

ohe_encoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
X_train_ohe_encoded = ohe_encoder.fit_transform(X_train_ohe)
X_test_ohe_encoded = ohe_encoder.transform(X_test_ohe)

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø–æ—Ä—è–¥–∫–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
ord_imputer = SimpleImputer(strategy='most_frequent')
X_train_ord = ord_imputer.fit_transform(X_train[ord_columns])
X_test_ord = ord_imputer.transform(X_test[ord_columns])

ord_encoder = OrdinalEncoder(categories=[['–ø—Ä–µ–º–∏—É–º', '—Å—Ç–∞–Ω–¥–∞—Ä—Ç']], handle_unknown='use_encoded_value', unknown_value=np.nan)
X_train_ord_encoded = ord_encoder.fit_transform(X_train_ord)
X_test_ord_encoded = ord_encoder.transform(X_test_ord)

# –û–±—Ä–∞–±–æ—Ç–∫–∞ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
num_scaler = MinMaxScaler()
X_train_num_scaled = num_scaler.fit_transform(X_train[num_columns])
X_test_num_scaled = num_scaler.transform(X_test[num_columns])

X_train_processed = np.hstack((X_train_ohe_encoded, X_train_ord_encoded, X_train_num_scaled, X_train.drop(ohe_columns + ord_columns + num_columns, axis=1)))
X_test_processed = np.hstack((X_test_ohe_encoded, X_test_ord_encoded, X_test_num_scaled, X_test.drop(ohe_columns + ord_columns + num_columns, axis=1)))
X_train_processed = pd.DataFrame(
    X_train_processed,
    columns=(
        list(ohe_encoder.get_feature_names(ohe_columns)) +
        ord_columns +
        num_columns +
        list(X_train.drop(ohe_columns + ord_columns + num_columns, axis=1).columns)
    )
)

# –°–æ–∑–¥–∞–Ω–∏–µ DataFrame –∏–∑ X_test_processed
X_test_processed = pd.DataFrame(
    X_test_processed,
    columns=(
        list(ohe_encoder.get_feature_names(ohe_columns)) +
        ord_columns +
        num_columns +
        list(X_test.drop(ohe_columns + ord_columns + num_columns, axis=1).columns)
    )
)
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ SVC —Å –ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
best_svc_model = SVC(C=1, probability=True, random_state=RANDOM_STATE)
best_svc_model.fit(X_train_processed, y_train)

# –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –∏ –º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_test_pred_proba = best_svc_model.predict_proba(X_test_processed)[:, 1]
roc_auc = roc_auc_score(y_test, y_test_pred_proba)
print('–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:', roc_auc)
```

    .best_estimator_.named_steps['models']
